## You must pick one indicator out of the four: population, life expectancy, GDP per capita, or suiside rate

```python
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')

df3
```

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 Population of Australia, Norway, Sweden and Turkiye", 
    color="Country Name"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 Population of Australia, Norway, Sweden and Turkiye", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

## You must compare all regions to see how they differ in the indicator

*   Compare one year (pick the latest year) using bar chart, pie chartList item




```python
pop1 = df.loc[df['indicator']=='SP.POP.TOTL']
pop1.head()
```

```python
population = pop1.loc[pop1['Year'] == 2021]
population
```

```python
df5 = population.groupby('Region')['value'].mean()
df5
```

```python
fig=px.bar(df5,x=df5.index,y='value', color=df5.index,title='2021 Population of Australia, Norway, Sweden and Turkiye')
fig.show()
```

```python
fig = px.pie(df5, values='value', names=df5.index, title='2021 Population of Australia, Norway, Sweden and Turkiye')
fig.show()
```

### Compare all years using Line Chart

```python
mean_pop=pop1.groupby(['Region','Year'])['value'].mean()
mean_pop=mean_pop.reset_index()
```

```python
fig=px.line(mean_pop,x='Year',y='value',color='Region',title='2021 Mean Population of Australia, Norway, Sweden and Turkiye')
fig.show()
```
